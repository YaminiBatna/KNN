```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
error = []
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

n_samples_train = X_train.shape[0]

# Ensure n_neighbors is less than or equal to n_samples_train
for i in range(1, min(40, n_samples_train + 1)):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    pred_i = model.predict(X_test)
    error.append(np.mean(pred_i != y_test))

plt.figure(figsize=(12, 6))
plt.plot(range(1, min(40, n_samples_train + 1)), error, color='red', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3 , metric = 'jaccard')
model.fit(X_train, y_train)
```

```python
clump_thickness = int(input("Enter Clump Thickness: "))
uniformity_cell_size = int(input("Enter Uniformity of Cell Size: "))
uniformity_cell_shape = int(input("Enter Uniformity of Cell Shape: "))
marginal_adhesion = int(input("Enter Marginal Adhesion: "))
single_epithelial_size = int(input("Enter Single Epithelial Cell Size: "))
bare_nuclei = int(input("Enter Bare Nuclei: "))
bland_chromatin = int(input("Enter Bland Chromatin: "))
normal_nucleoli = int(input("Enter Normal Nucleoli: "))
mitoses = int(input("Enter Mitoses: "))
newcancer = [[clump_thickness, uniformity_cell_size, uniformity_cell_shape, marginal_adhesion, single_epithelial_size, bare_nuclei, bland_chromatin, normal_nucleoli, mitoses]]
result = model.predict(sc.transform(newcancer))
print(result)
if result == 1:
  print("Patient has Cancer (Malignant)")
else:
  print("Patient has no Cancer (Benign)")
```
